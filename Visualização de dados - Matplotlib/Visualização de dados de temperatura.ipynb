Customizando as visualizações

Mudando espessura da linha com o atributo lw em eixo.plot

```python
figura = plt.figure(figsize=(15,8))
eixo = figura.add_axes([0, 0, 1, 1])
eixo.plot(dados.data, dados.temperatura, color = 'g', lw=4)

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

```python
figura = plt.figure(figsize=(15,8))
eixo = figura.add_axes([0, 0, 1, 1])
eixo.plot(dados.data, dados.temperatura, color = 'g', lw=0.5)

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

Alterando o estilo de linha com o atributo ls

```python
figura = plt.figure(figsize=(15,8))
eixo = figura.add_axes([0, 0, 1, 1])
eixo.plot(dados.data, dados.temperatura, color = 'g', ls='dotted')

eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

Limitando o período do gráfico 2014 até 2015 utilizando o método .xlim e a lib datetime

```python
import datetime
```

```python
figura = plt.figure(figsize=(15,8))
eixo = figura.add_axes([0, 0, 1, 1])
eixo.plot(dados.data, dados.temperatura, color = 'g')

eixo.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo.set_title('Temperatura no momento', fontsize=25)
eixo.set_ylabel('Temperatura (K)', fontsize=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.legend(['Temperatura'], loc = 'lower right', fontsize=15)
```

Pontos máximo e mínimo

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados.data, dados.temperatura, color='g')
eixo.set_title('Temperatura 2012-2018', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura(K)', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

eixo.axhline(max(dados.temperatura), color='red', linestyle='--')
eixo.axhline(min(dados.temperatura), color='red', linestyle='--')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados.data, dados.temperatura, color='g')
eixo.set_title('Temperatura 2012-2018', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura(K)', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados.temperatura)

eixo.annotate('Máximo', xy=(x1,y1), fontsize=20)

eixo.axhline(max(dados.temperatura), color='red', linestyle='--')
eixo.axhline(min(dados.temperatura), color='red', linestyle='--')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados.data, dados.temperatura, color='g')
eixo.set_title('Temperatura 2012-2018', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura(K)', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados.temperatura)

x2 = dados['data'][dados['temperatura'].idxmax() - 7000]
y2 = max(dados.temperatura) - 5

eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2),
             arrowprops = dict(facecolor='k'))

eixo.axhline(max(dados.temperatura), color='red', linestyle='--')
eixo.axhline(min(dados.temperatura), color='red', linestyle='--')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.plot(dados.data, dados.temperatura, color='g')
eixo.set_title('Temperatura 2012-2018', fontsize=25, pad=20)
eixo.set_xlabel('Data', fontsize=20)
eixo.set_ylabel('Temperatura(K)', fontsize=20)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.grid(True)

x1 = dados['data'][dados['temperatura'].idxmax()]
y1 = max(dados.temperatura)
x2 = dados['data'][dados['temperatura'].idxmax() - 7000]
y2 = max(dados.temperatura) - 5
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2),
             arrowprops = dict(facecolor='k'))

x1 = dados['data'][dados['temperatura'].idxmin()]
y1 = min(dados.temperatura)
x2 = dados['data'][dados['temperatura'].idxmin() - 7000]
y2 = min(dados.temperatura) + 5
eixo.annotate('Máximo', xy=(x1,y1), fontsize=20, xytext=(x2,y2),
             arrowprops = dict(facecolor='k'))

eixo.axhline(max(dados.temperatura), color='red', linestyle='--')
eixo.axhline(min(dados.temperatura), color='red', linestyle='--')
```

```python
temp_por_dia_da_semana = dados.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temp_por_dia_da_semana = temp_por_dia_da_semana[nome_dias]
temp_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))

eixo.bar(indice, temp_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=10)
eixo.set_ylabel('Temperatura(K)', fontsize=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))

eixo.bar(indice, temp_por_dia_da_semana)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=10)
eixo.set_ylabel('Temperatura(K)', fontsize=10)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temp_por_dia_da_semana, color=cores)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=10)
eixo.set_ylabel('Temperatura(K)', fontsize=10)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temp_por_dia_da_semana))
cores = ['black', 'r', 'g', 'b', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temp_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
eixo.set_xlabel('Dia da semana', fontsize=10)
eixo.set_ylabel('Temperatura(K)', fontsize=10)
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%')
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [1,0,0,0,0,1,1]

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [.1,0,0,0,0,.1,.1]

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir)
eixo.set_title('Temperatura por dia da semana', fontsize=15, pad=10)
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

explodir = [.1,0,0,0,0,.1,.1]

eixo.pie(temp_por_dia_da_semana, labels=temp_por_dia_da_semana.index,
        autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura média por dia da semana', fontsize=15, pad=10)
```
